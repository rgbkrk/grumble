```python
def hsl(hue=120, saturation=1, luminance=1):
    return f"hsl({hue}, {saturation*100}%, {luminance*100}%)"
```

```python
def walk(val=0.5, step=0.02):
    delta = np.random.uniform(-step, step)
    return max(min(val + delta, 1), 0)
```

```python
def rgb(r, g, b):
    return f'rgb({int(r*256)},{int(g*256)},{int(b*256)})'

def block(color='DeepPink', size=20):
    return div(
        style={
            'width': f'{size}px',
            'height': f'{size}px',
            'backgroundColor': color,
            'display': 'inline-block'
        }
    )
```

```python
div(
    [block(color) for color in ['red', 'orange', 'yellow', 'green', 'blue', 'purple']]
)
```

```python
def show_colormap(cm=matplotlib.cm.viridis):
    return div(
        code(cm.name, style={'width': '10vw', 'font-weight': '600'}),
        [block(color) for color in [rgb(*row) for row in cm.colors]],
        style={
            'display': 'flex',
            'paddingBottom': '10px'
        }
    )
```

```python
div(
    [show_colormap(y) for (x,y) in matplotlib.cm.__dict__.items() if hasattr(y, 'colors') and hasattr(y, 'name')]
)
```
